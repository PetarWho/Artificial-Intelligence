# Fraud Detection in Financial Transactions Using Anomaly Detection Techniques

## Background
In the age of digital commerce, credit card fraud poses significant financial and reputational risks. Credit card companies strive to identify fraudulent transactions to protect their customers and reduce losses. Fraudulent activities are often rare, making the dataset highly imbalanced, and complicating the task of designing an effective detection system.

## Dataset Overview
This project uses the Credit Card Fraud dataset from Kaggle, which contains anonymized transactions from European cardholders over two days. 

The dataset includes:
- 284,807 total transactions.
- 492 fraudulent transactions (0.172%), highlighting its imbalanced nature.
- 30 features, with most being principal components from a PCA transformation to ensure confidentiality. The features include:
    - Time: Seconds elapsed since the first transaction.
    - Amount: The transaction value.
    - Class: Target variable, where 1 indicates fraud, and 0 indicates legitimate transactions.
    - The anonymized nature of the data and class imbalance present unique challenges in designing a detection model.

## Objective
The objective of this project is to build a system to detect fraudulent transactions using anomaly detection techniques such as:
- Random Forest
- XGBoost

## Significance
This project addresses a real-world problem of fraud detection with methods that generalize to unseen data, offering practical applications in fraud prevention systems. Anomaly detection techniques are chosen for their ability to identify rare events.

### Random Forest Algorithm

The Random Forest Algorithm is a **supervised learning model** used for both **regression** and **classification** problems. It builds multiple decision trees during training and combines their outputs for improved accuracy and robustness.

---

#### Pros and Cons
**Pros**:
- Handles both regression and classification tasks effectively.
- Reduces the risk of overfitting due to averaging of multiple trees.
- Fast training process.

**Cons**:
- Predictions can be slow for large datasets due to multiple tree evaluations.
- Sensitive to outliers and missing data.

---

### Mathematical Concepts

#### Regression Problems
For regression tasks, Random Forest minimizes the **[Mean Squared Error (MSE)](https://en.wikipedia.org/wiki/Mean_squared_error)** at each node:
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (f_i - y_i)^2
$$

Where:
- $y_i$: Actual value of the data point.
- $f_i$: Predicted value from the decision tree.
- $n$: Total number of data points.

#### Classification Problems
For classification tasks, Random Forest evaluates split quality using the **[Gini Index](https://en.wikipedia.org/wiki/Gini_coefficient)**:
$$
\text{Gini} = 1 - \sum_{i=1}^c (p_i)^2
$$

Where:
- $p_i$: Proportion of observations belonging to class $i$.
- $c$: Total number of classes.

Alternatively, it can use **[Entropy](https://en.wikipedia.org/wiki/Entropy_(information_theory))**:
$$
\text{Entropy} = - \sum_{i=1}^c p_i \log_2(p_i)
$$
Entropy is more computationally intensive due to the logarithmic function.

---

#### Summary
Random Forest is a versatile algorithm that excels in diverse applications but requires attention to outliers and data quality for optimal performance.


```python
from sklearn.ensemble import RandomForestClassifier

def print_random_forest_evaluation(X_train, Y_train, X_test, Y_test):
    rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
    rf_model.fit(X_train, Y_train)
    
    Y_pred = rf_model.predict(X_test)
    Y_proba = rf_model.predict_proba(X_test)[:, 1]
    
    print("=== Classification Report ===")
    print(classification_report(Y_test, Y_pred))
    
    print("\n=== Accuracy Score ===")
    print(f"Accuracy: {accuracy_score(Y_test, Y_pred):.4f}")
    
    print("\n=== AUC-ROC Score ===")
    print(f"AUC-ROC: {roc_auc_score(Y_test, Y_proba):.4f}")


print_random_forest_evaluation(X_train, Y_train, X_test, Y_test)
```

### XGBoost: Overview and Mathematics

XGBoost is a powerful implementation of Gradient Boosted Trees (GBT), a supervised learning algorithm. It optimizes a specified loss function and incorporates regularization techniques to improve generalization. It is widely used for both classification and regression tasks due to its efficiency and accuracy.

The **objective function** of XGBoost is composed of two parts:
1. **Loss function** $l$: Measures the error between predictions and true values.
2. **Regularization** $\Omega$: Controls model complexity.

---

#### Function

The XGBoost objective function at iteration $t$ is defined as:

$$
\text{Objective} = \sum_{i=1}^n l(y_i, \hat{y}_i^{(t-1)} + f_t(x_i)) + \Omega(f_t)
$$

where:
- $y_i$: True labels.
- $\hat{y}_i^{(t-1)}$: Predictions from the model up to step $t-1$.
- $f_t$: The new learner (CART tree) added at step $t$.
- $\Omega(f_t) = \gamma T + \frac{1}{2} \lambda \sum_{j=1}^T w_j^2$: Regularization term.
  - $T$: Number of leaves.
  - $w_j$: Leaf weights.
  - $\gamma$: Regularization on the number of leaves.
  - $\lambda$: L2 regularization parameter.

---

#### Taylor Approximation for Optimization

Since the objective is a sum of functions of functions, traditional optimization techniques cannot be applied directly. Using the **second-order Taylor expansion**, the objective can be approximated as:

$$
\text{Objective}_t = \sum_{i=1}^n \left[ g_i f_t(x_i) + \frac{1}{2} h_i f_t^2(x_i) \right] + \Omega(f_t)
$$

where:
- $g_i = \frac{\partial l(y_i, \hat{y}_i)}{\partial \hat{y}_i}$: Gradient of the loss.
- $h_i = \frac{\partial^2 l(y_i, \hat{y}_i)}{\partial \hat{y}_i^2}$: Hessian (second derivative) of the loss.

By removing constant terms, the objective simplifies to:

$$
\text{Simplified Objective}_t = \sum_{j=1}^T \left[ G_j w_j + \frac{1}{2} H_j w_j^2 \right] + \gamma T + \frac{1}{2} \lambda \sum_{j=1}^T w_j^2
$$

where:
- $G_j = \sum_{i \in \text{leaf}_j} g_i$: Sum of gradients for leaf $j$.
- $H_j = \sum_{i \in \text{leaf}_j} h_i$: Sum of Hessians for leaf $j$.

The optimal weight for leaf $j$ is:

$$
w_j^* = -\frac{G_j}{H_j + \lambda}
$$

---

#### Scoring Function for Splits

To evaluate splits during tree construction, the **gain** (reduction in loss) is calculated as:

$$
\text{Gain} = \frac{1}{2} \left[ \frac{G_L^2}{H_L + \lambda} + \frac{G_R^2}{H_R + \lambda} - \frac{(G_L + G_R)^2}{H_L + H_R + \lambda} \right] - \gamma
$$

where:
- $G_L, G_R$: Gradients for left and right branches.
- $H_L, H_R$: Hessians for left and right branches.

If the gain is positive and exceeds a minimum threshold $\text{min\_split\_gain}$, the branch is split further.

---

#### Binary Classification with Log Loss

For binary classification, the **log loss** is used as the objective:

$$
l(y, \hat{p}) = -\left[ y \log(\hat{p}) + (1-y) \log(1-\hat{p}) \right]
$$

Here:
- $y \in \{0, 1\}$: True label.
- $\hat{p}$: Predicted probability after applying the sigmoid function:

$$
\hat{p} = \sigma(x) = \frac{1}{1 + e^{-x}}
$$

The gradient and Hessian for log loss are:

$$
g = \hat{p} - y, \quad h = \hat{p}(1-\hat{p})
$$

---

#### Summary of the Algorithm

1. **Initialization**: Start with a constant prediction.
2. **Iterative Learning**: At each iteration $t$:
   - Compute gradients $g_i$ and Hessians $h_i$.
   - Construct a CART tree using the gain function.
   - Update predictions with the new tree.
3. **Stopping Criteria**: Stop when further iterations do not significantly reduce the loss.

```python
from xgboost import XGBClassifier

def print_xgboost_evaluation(X_train, Y_train, X_test, Y_test):
    xgb_model = XGBClassifier(scale_pos_weight=len(Y_train[Y_train == 0]) / len(Y_train[Y_train == 1]),
                              eval_metric='auc', random_state=42)
    xgb_model.fit(X_train, Y_train)
    
    Y_pred = xgb_model.predict(X_test)

    Y_proba = xgb_model.predict_proba(X_test)[:, 1]
    
    print("=== Classification Report ===")
    print(classification_report(Y_test, Y_pred))
    
    print("\n=== Accuracy Score ===")
    print(f"Accuracy: {accuracy_score(Y_test, Y_pred):.4f}")
    
    print("\n=== AUC-ROC Score ===")
    print(f"AUC-ROC: {roc_auc_score(Y_test, Y_proba):.4f}")


print_xgboost_evaluation(X_train, Y_train, X_test, Y_test)
```
